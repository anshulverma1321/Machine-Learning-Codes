```python
import numpy as np
from scipy import stats
import pandas as pd
```

```python
scores = [85, 78, 92, 88, 75, 95, 89, 84, 90, 77]
```

```python
np.median(scores)
```

```python
np.mean(scores)
```

```python
mode_result = stats.mode(scores, keepdims=True)
print("Mode:", mode_result.mode[0])
```

```python
q3 = np.percentile(scores, 75)
q1 = np.percentile(scores, 25)
iqr = q3 - q1
print("IQR:", iqr)
```
